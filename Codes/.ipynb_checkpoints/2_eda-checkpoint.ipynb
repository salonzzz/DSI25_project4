The chart above shows that there is a sharp decline in the number of samples with Wnv present from 2007 to 2009. From 2009 to 2013, the number of samples with Wnv present had increased. 

A possible reason for the higher number of samples with Wnv present in 2007 could be because of the higher mosquito breedings, ceteris paribus. However, assuming that the level of mosquito breedings do not change significantly over the years, a plausible explanation for the trend could be due to changes in the sampling efforts of the public health authorities in Chicago (i.e. reduction in sampling efforts, could reduce the absolute number of samples with Wnv present). We will corroborate this by charting the number of mosquitoes and sampling efforts over time later.

### Kernal Density Heatmap

```python
#Define a function that takes in data 
def plotHeatMap(data, a_cm,ax):
    if(len(data) > 0):
        X = data[['Longitude', 'Latitude']].values
        kd = KernelDensity(bandwidth=0.02)
        kd.fit(X)
        xv,yv = np.meshgrid(np.linspace(-88, -87.5, 100), np.linspace(41.6, 42.1, 100))
        gridpoints = np.array([xv.ravel(),yv.ravel()]).T
        zv = np.exp(kd.score_samples(gridpoints).reshape(100,100))
        ax.imshow(zv, 
                   origin='lower', 
                   cmap=a_cm, 
                   extent=lon_lat_box, 
                   aspect=aspect)
```

```python
# load map
mapdata = np.loadtxt('https://raw.githubusercontent.com/zixinlee/GA-data-science/master/West-nile-virus/data/mapdata_copyright_openstreetmap_contributors.txt')

# heatmap settings
alpha_wnv = plt.cm.YlOrBr
alpha_wnv._init()
alpha_wnv._lut[:-3,-1] = abs(np.logspace(0, 1, alpha_wnv.N) / 10 - 1)[::-1]
alpha_sp = plt.cm.Greens
alpha_sp._init()
alpha_sp._lut[:-3,-1] = abs(np.logspace(0, 1, alpha_wnv.N) / 10 - 1)[::-1]

# setting boundaries for map
aspect = mapdata.shape[0] * 1.0 / mapdata.shape[1]
lon_lat_box = (-88, -87.5, 41.6, 42.1)

# create plot for each month and year 
#loop year
for year in df_train['Date'].dt.year.unique():
    fig, ax = plt.subplots(ncols = 6, figsize = (32,10))
    
    #loop month
    for i, month in enumerate(df_train['Date'].dt.month.unique()):
        # plot basemap
        ax[i].imshow(mapdata, cmap=plt.get_cmap('gray'), extent=lon_lat_box, aspect=aspect)
        current_plot = ax.ravel()
        
        # plot station locations
        station = np.array([[-87.933, 41.995], [-87.752, 41.786]])
        ax[i].scatter(station[:,0], station[:,1], marker='o', c = 'blue', s = 50, alpha = 0.7, label = 'Station')
        
        # plot wnv locations
        wnv = df_train[(df_train['Date'].dt.year == year) & (df_train['Date'].dt.month == month) & (df_train['WnvPresent'] == 1)]
        wnv = wnv.groupby(['Date','Trap', 'Longitude', 'Latitude']).max()['WnvPresent'].reset_index()
        plotHeatMap(wnv, alpha_wnv,ax[i])
        
        if year==2011:
           # include spray location
           df_spray__2011 = df_spray[(df_spray['Date'].dt.year==2011) & (df_spray['Date'].dt.month == month)]
           df_spray__2011 = df_spray__2011.groupby(['Date','Longitude', 'Latitude']).max().reset_index()
           plotHeatMap(df_spray__2011, alpha_sp,ax[i])
           
        if year==2013:
           # include spray location
           df_spray__2013 = df_spray[(df_spray['Date'].dt.year==2013) & (df_spray['Date'].dt.month == month)]
           df_spray__2013 = df_spray__2013.groupby(['Date','Longitude', 'Latitude']).max().reset_index()
           plotHeatMap(df_spray__2013, alpha_sp,ax[i])
        
        # plot trap locations
        locations = df_train[['Longitude', 'Latitude']].drop_duplicates().values
        ax[i].scatter(locations[:,0], locations[:,1], marker='x', c = 'black', s = 20, alpha = 1, label = 'Trap')
        
        
        # title and axes
        ax[i].set_title(f"Year({year}) and Month({month})", fontsize = 15, ha = 'left', position = (0,1))
        ax[i].legend()
```

The Kernal Density heatmap provides us with a density approximation of the presence of the west nile virus based on the number of points in a given location (or in this case, at a given latitude and longitude), with larger numbers of clustered points resulting in larger values - and hence denser outputs. The same concept is applied to the spray dataset which turns up as green highlights in the above maps.

The maps depict the trap locations (black), station locations (blue), presence and estimated density of west nile virus (red) and presence and estimated density of the anti-mosquito spray (green) across 6 months (May-October) in 2007, 2009,2011 and 2013.

As we move from month 5 (May) to month 10 (October) in each year, we see a peak in the occurrence of the west nile virus between months 7-9 (July-September). This is with the exception of 2007, in which the virus continued to be detected in October and 2013, in which the presence of the virus in detected earlier, in June. 

Sprays in 2011 were only conducted in September while sprays in 2013 were conducted across 3 months from July to September. 

Interestingly we observe that in 2009,2011 and 2013, no presence of the west nile virus was detected in October. In the case of 2011 and 2013, this could potentially be attributed to the sprays however there might be other environmental variables i.e. weather, at play that could be influencing the uptick in the presence and detection of the west nile virus.

```python
#group by date and trap, ignoring species
df_date_trap = df_combined[['Date','Trap','NumMosquitos','WnvPresent']].groupby(by=['Date','Trap']).sum().reset_index()
df_date_trap['WnvPresent'] = df_date_trap['WnvPresent'].map(lambda x: 1 if x > 0 else 0)
df_date_trap['NumTraps'] = 1
df_date_trap.head()
```

```python
#calculate proportion of traps that are positive for each date
df_date_trap_proportion = df_date_trap.groupby(by='Date').sum().reset_index()
df_date_trap_proportion['ProportionTrapsPos'] = df_date_trap_proportion['WnvPresent']/df_date_trap_proportion['NumTraps']
df_date_trap_proportion['Year'] = df_date_trap_proportion['Date'].apply(lambda x: x.year)
df_date_trap_proportion.sort_values(by = 'ProportionTrapsPos', ascending = False).head()
```

Dates in 2007 appeared to have the highest proportion of traps, but taking a closer look, many of these dates only have a small number of traps that were sampled and might not reflect the true prevalence of WNV across the country

```python
#check distribution of number of traps sampled per day, for each year
fig, ax = plt.subplots(1,1,figsize=(12,8))
sns.boxplot(x='Year', y='NumTraps', data=df_date_trap_proportion, color = 'lightblue', showfliers = True, ax=ax).\
set_title('Distribution of no. of traps sampled daily, by year');
```

There is quite a wide range of number of traps sampled daily in 2007 and 2009, hence it may not be appropriate to look at proportion of traps that tested positive for Wnv daily for those years

```python
#check distribution of number of mosquitos caught per day, for each year
fig, ax = plt.subplots(1,1,figsize=(12,8))
sns.boxplot(x='Year', y='NumMosquitos', data=df_date_trap_proportion, color = 'lightblue', showfliers = True, ax=ax).\
set_title('Distribution of no. of mosquitos caught daily, by year');
```

There are some outliers in 2007, with an exceptionally high number of mosquitos captured for certain days
